CLOSING PRICE

```python
import matplotlib.ticker as ticker

def closing_price(df, column_name):
    # Melakukan pivoting untuk mendapatkan data harga penutup yang disesuaikan (Adj Close) masing-masing bank
    pivot_df = df.pivot(index='Date', columns='Bank_Name', values=column_name)

    # Membuat subplot terpisah untuk setiap bank
    fig, axes = plt.subplots(len(pivot_df.columns), 1, figsize=(15, 8), sharex=True)

    # Loop melalui setiap bank menggunakan enumerate
    for i, (bank, ax) in enumerate(zip(pivot_df.columns, axes), 1):
        ax.plot(pivot_df.index, pivot_df[bank], label=bank)
        ax.set_ylabel(f'{column_name} Price (IDR)')  # Menambahkan satuan rupiah

        # Menambahkan label harga tertinggi dan terendah pada setiap subplot
        ax.annotate(f'Highest: {pivot_df[bank].max():,.2f} IDR', 
                    xy=(pivot_df.idxmax()[bank], pivot_df[bank].max()), 
                    xytext=(10, -20),
                    textcoords='offset points',
                    arrowprops=dict(facecolor='green', arrowstyle='wedge,tail_width=0.7', alpha=0.5),
                    bbox=dict(boxstyle='round,pad=0.3', edgecolor='red', facecolor='white', alpha=0.5))

        ax.annotate(f'Lowest: {pivot_df[bank].min():,.2f} IDR', 
                    xy=(pivot_df.idxmin()[bank], pivot_df[bank].min()), 
                    xytext=(10, 20),
                    textcoords='offset points',
                    arrowprops=dict(facecolor='red', arrowstyle='wedge,tail_width=0.7', alpha=0.5),
                    bbox=dict(boxstyle='round,pad=0.3', edgecolor='green', facecolor='white', alpha=0.5))

        # Menambahkan informasi rata-rata
        avg_price = pivot_df[bank].mean()
        ax.axhline(y=avg_price, color='blue', linestyle='--', label=f'Average: {avg_price:,.2f} IDR')
        ax.annotate(f'Avg: {avg_price:,.2f} IDR', 
                    xy=(pivot_df.index[0], avg_price), 
                    xytext=(10, 0),
                    textcoords='offset points',
                    color='blue')

        # Menggunakan formatter untuk format mata uang IDR
        ax.get_yaxis().set_major_formatter(ticker.StrMethodFormatter('{x:,.0f} IDR'))

        ax.set_title(f'Closing Price Over Time of {bank}')

    # Menambahkan label dan judul untuk sumbu x
    axes[-1].set_xlabel('Date')

    # Menata layout
    plt.tight_layout()
    plt.show()

    # Membuat plot perbandingan harga penutup yang disesuaikan
    plt.figure(figsize=(15, 8))
    for bank in pivot_df.columns:
        plt.plot(pivot_df.index, pivot_df[bank], label=bank)

    # Menambahkan label dan judul
    plt.xlabel('Date')
    plt.ylabel(f'{column_name} Price (IDR)')  # Menambahkan satuan rupiah
    plt.title(f'Comparison of {column_name} Prices for Different Banks')
    
    # Menambahkan informasi rata-rata pada plot keseluruhan
    plt.axhline(y=pivot_df.mean().mean(), color='blue', linestyle='--', label=f'Overall Average: {pivot_df.mean().mean():,.2f} IDR')
    plt.annotate(f'Overall Avg: {pivot_df.mean().mean():,.2f} IDR', 
                xy=(pivot_df.index[0], pivot_df.mean().mean()), 
                xytext=(10, 5),
                textcoords='offset points',
                color='blue')

    # Menggunakan formatter untuk format mata uang IDR pada sumbu y
    plt.gca().yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f} IDR'))

    plt.legend()
    plt.show()

# Gunakan fungsi untuk plot harga penutup yang disesuaikan (Adj Close)
closing_price(df, 'Adj Close')
```

VOLUME OF SALES

```python
def plot_comparison(df, column_name):
    # Melakukan pivoting untuk mendapatkan data harga penutup yang disesuaikan (Adj Close) masing-masing bank
    pivot_df = df.pivot(index='Date', columns='Bank_Name', values=column_name)

    # Membuat subplot terpisah untuk setiap bank
    fig, axes = plt.subplots(len(pivot_df.columns), 1, figsize=(15, 8), sharex=True)

    # Loop melalui setiap bank menggunakan enumerate
    for i, (bank, ax) in enumerate(zip(pivot_df.columns, axes), 1):
        ax.plot(pivot_df.index, pivot_df[bank], label=bank)
        ax.set_ylabel(f'{column_name} Value')
        ax.set_title(f'{column_name} Over Time of {bank}')

        # Menambahkan label nilai tertinggi dan terendah pada setiap subplot
        ax.annotate(f'Highest: {pivot_df[bank].max():,.2f}', 
                    xy=(pivot_df.idxmax()[bank], pivot_df[bank].max()), 
                    xytext=(10, -20),
                    textcoords='offset points',
                    arrowprops=dict(facecolor='red', arrowstyle='wedge,tail_width=0.7', alpha=0.5),
                    bbox=dict(boxstyle='round,pad=0.3', edgecolor='red', facecolor='white', alpha=0.5))

        ax.annotate(f'Lowest: {pivot_df[bank].min():,.2f}', 
                    xy=(pivot_df.idxmin()[bank], pivot_df[bank].min()), 
                    xytext=(10, 20),
                    textcoords='offset points',
                    arrowprops=dict(facecolor='green', arrowstyle='wedge,tail_width=0.7', alpha=0.5),
                    bbox=dict(boxstyle='round,pad=0.3', edgecolor='green', facecolor='white', alpha=0.5))

        # Menggunakan formatter untuk format mata uang IDR pada sumbu y
        ax.get_yaxis().set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))

    # Menambahkan label dan judul untuk sumbu x
    axes[-1].set_xlabel('Date')

    # Menata layout
    plt.tight_layout()
    plt.show()

    # Membuat plot perbandingan harga penutup yang disesuaikan
    plt.figure(figsize=(15, 8))
    for bank in pivot_df.columns:
        plt.plot(pivot_df.index, pivot_df[bank], label=bank)

    # Menambahkan garis harga tertinggi dan terendah pada plot keseluruhan
    plt.axhline(y=pivot_df.max().max(), color='r', linestyle='--', label=f'Highest {column_name}')
    plt.axhline(y=pivot_df.min().min(), color='g', linestyle='--', label=f'Lowest {column_name}')

    # Menambahkan label dan judul
    plt.xlabel('Date')
    plt.ylabel(f'{column_name} Value')
    plt.title(f'Comparison of {column_name} Values for Different Banks')
    plt.legend()
    plt.show()

# Gunakan fungsi untuk plot volume
plot_comparison(df, 'Volume')
```

GROWTH OF STOCK

```python
import numpy as np

# Assuming df is the DataFrame containing the stock data

# Find the minimum and maximum closing prices for each bank
mini = [df[df['Date'] == df['Date'].min()]['Close'].values.item() for name, df in df.groupby('Bank_Name')]
maxi = [df[df['Date'] == df['Date'].max()]['Close'].values.item() for name, df in df.groupby('Bank_Name')]

# Calculate the absolute difference between the closing prices
diff = np.array(maxi) - np.array(mini)

# Calculate the percentage growth
growth = (diff / np.array(mini)) * 100

# Convert growth to a list
growth_list = growth.tolist()

# Get the unique bank names
bank_names = df['Bank_Name'].unique()

# Create a dictionary to map bank names to their corresponding growth
growth_dict = dict(zip(bank_names, growth_list))

print("Percentage Growth:")
for bank_name, growth in growth_dict.items():
    print(f"{bank_name}: {growth:.2f}%")
```

```python
plt.figure(figsize=(10, 6))
plt.bar(growth_dict.keys(), growth_dict.values(), color=['green' if g > 0 else 'red' for g in growth_dict.values()])
plt.xlabel('Bank Name')
plt.ylabel('Percentage Growth')
plt.title('Percentage Growth of Banks')
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')  # Add a horizontal line at 0 for reference

# Add data labels
for bank_name, growth in growth_dict.items():
    plt.text(bank_name, growth, f"{growth:.2f}%", ha='center', va='bottom' if growth > 0 else 'top')

# Show the plot
plt.tight_layout()
plt.show()
```

DAILY RETURN

```python
def daily_return_plot(df):
    # Calculate daily return
    df['Daily Return'] = df.groupby('Bank_Name')['Close'].pct_change() * 100

    # Assign variables to annotation positions
    ave_x = df['Date'].mean()
    y_max = df['Daily Return'].max()
    y_max_date = df[df['Daily Return'] == y_max]['Date'].values[0]
    xb = pd.to_datetime(y_max_date).date()
    y_min = df['Daily Return'].min()
    y_mean = df['Daily Return'].mean()

    # Plotting
    plt.figure(figsize=(13, 6), facecolor='none')  # Set facecolor to 'none' for transparent background

    plt.subplot(121)
    plt.plot(df['Date'], df['Daily Return'], color='#062e63')
    plt.axhline(y=y_max, color='green', ls='--')
    plt.axhline(y=y_min, color='red', ls='--')
    plt.xlabel('Date')
    plt.ylabel("Percentage return")
    plt.annotate(f"Min. daily return={round(y_min, 2)}%",
                 xy=(ave_x, y_min), xytext=(ave_x, y_min),
                 bbox=dict(boxstyle="round", facecolor='none', edgecolor='#d0d5db')  # Set facecolor to 'none'
                 )
    plt.annotate(f"Max. daily return={round(y_max, 2)}%\nDate was={xb}",
                 xy=(ave_x, y_max), xytext=(ave_x, y_max - 0.6),
                 bbox=dict(boxstyle="round", facecolor='none', edgecolor='#d0d5db', alpha=0.7)  # Set facecolor to 'none'
                 )
    plt.title(f"Daily return of stock {df['Bank_Name'].unique()[0]}")

    plt.subplot(122)
    plt.hist(df['Daily Return'].dropna(), density=True, color='#0f52a8')
    plt.xlabel('Histogram')
    plt.axvline(x=df['Daily Return'].mean(), color='yellow', ls='--')
    plt.annotate(f"Mean daily return={round(y_mean, 2)}%",
                 xy=(y_mean, 0.10), xytext=(y_mean + 2, 0.10),
                 bbox=dict(boxstyle="round", facecolor='none', edgecolor='#d0d5db'),  # Set facecolor to 'none'
                 arrowprops=dict(arrowstyle="wedge,tail_width=1.", facecolor='none',
                                 edgecolor='#d0d5db',
                                 relpos=(0.1, 0.5)
                                 )
                 )
    plt.title(f"Histogram of stock price")

    plt.show()

# Example usage
# Assuming 'df' is your DataFrame
daily_return_plot(df[df['Bank_Name'] == 'BCA'])
daily_return_plot(df[df['Bank_Name'] == 'BNI'])
daily_return_plot(df[df['Bank_Name'] == 'BRI'])
```

candle stick

```python
import plotly.graph_objects as go
import pandas as pd

# Assuming your DataFrame is named 'df'
# If it's not, replace 'df' with the actual name of your DataFrame

# Convert 'Date' column to datetime type
df['Date'] = pd.to_datetime(df['Date'])

# Sort DataFrame by 'Date'
df.sort_values(by=['Bank_Name', 'Date'], inplace=True)

# Create separate candlestick charts for each bank
banks = df['Bank_Name'].unique()

for bank in banks:
    bank_data = df[df['Bank_Name'] == bank]

    fig = go.Figure(data=[go.Candlestick(x=bank_data['Date'],
                                         open=bank_data['Open'],
                                         high=bank_data['High'],
                                         low=bank_data['Low'],
                                         close=bank_data['Close'],
                                         name=bank)])

    # Update layout for better visibility
    fig.update_layout(title=f'Candlestick Chart - {bank} Stock Analysis',
                      xaxis_title='Date',
                      yaxis_title='Stock Price',
                      xaxis_rangeslider_visible=False)

    # Show the chart
    fig.show()
```

forecasting

```python
# Sort DataFrame by 'Date'
df.sort_values(by=['Bank_Name', 'Date'], inplace=True)

# List of unique banks in the DataFrame
banks = df['Bank_Name'].unique()

# Loop through each bank
for bank in banks:
    # Select the data for the current bank
    bank_data = df[df['Bank_Name'] == bank]

    # Use 'Close' prices for forecasting
    data = bank_data['Close']

    # Split the data into training and testing sets
    train_size = int(len(data) * 0.8)
    train, test = data[:train_size], data[train_size:]

    # Fit ARIMA model
    order = (5, 1, 1)  
    model = ARIMA(train, order=order)
    fit_model = model.fit()

    # Forecast
    forecast = fit_model.forecast(steps=len(test))

    # Plotting
    plt.figure(figsize=(10, 6))
    plt.plot(train.index, train, label='Training Data')
    plt.plot(test.index, test, label='Actual Prices')
    plt.plot(test.index, forecast, label='ARIMA Forecast', color='red')
    plt.title(f'ARIMA Forecasting for {bank} Stock Prices')
    plt.xlabel('Date')
    plt.ylabel('Stock Price')
    plt.legend()
    plt.show()

    # Calculate RMSE (Root Mean Squared Error)
    rmse = sqrt(mean_squared_error(test, forecast))
    print(f"Root Mean Squared Error (RMSE) for {bank}: {rmse}")
```
